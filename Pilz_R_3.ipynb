For this data analysis with python, the daily units sold are correlated wit the temperature. The higher the temperature, the higher the units sold per day. However, due to the positive correlation between the products themselves, it can be hypothesized that the higher sales stem from the seasonal climate changes in Australian, than the products themselves.

The File_path is through the google collaboration notebook. I uploaded the csv file into the notebook.
I will skip the first 2 rows by using the skiprows function.

```python
import pandas as pd
import os

file_path = '/content/daily_sales.csv'
df = pd.read_csv(file_path, skiprows = 2)
df.head()
```

The describe function gives an opportunity to see a sumarized table with count, mean, std, min, 25%, 50%, 75% and max.

```python
df.describe().round()
```

**Interpretation**

The Count column means that there are 244 instances of none null values in each column minus the 2 skipped rows.
The mean is the average of all values in each respective column.
The std shows the standard deviation in each column from the mean.
The min shows the minimum values in each respective column.
The 25%, 50% and the 75% gives a benchmark of what values lie under the respective percentages of the data.
the max shows the maximum value in each column.

```python
from scipy.stats import pearsonr

coeff = pearsonr(df['Daily Units Sold'], df['Daily Temperature (C)'])
print(f"\nPearson correlation coefficient between Daily Units Sold and Temperature: {coeff}")

```
